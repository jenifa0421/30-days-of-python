**1.Check that a string contains only a certain set of characters**

```
import re
def fun(text):
    r = re.compile(r'[^a-zA-Z0-9.]')
    x = r.search(text)
    return not bool(x)

print(fun("Je124")) 
print(fun("}{"))    
print(fun('10'))



```

**2. Write a Python program that matches a word containing 'ab'**

```
import re
def fun(text):
  pattern='ab'
  if re.search(pattern,text):
    return('match found')
  else:
    return('not found')
   
print(fun('First three letters of Alphabet is abc'))
print(fun('def'))
```

**3.Write a Python program to check for a number at the end of a string**

```
import re
def fun(string):
    r=re.compile(r".*[0-9]$")
    if r.match(string):
        return True
    else:
        return False

print(fun('jeni76'))
print(fun('keerthu92'))
print(fun('abcd'))
```

**4.Write a Python program to search the numbers (0-9) of length between 1 to 3 in a given string**

```
import re
pattern=r"([0-9]{1,3})"
t=" 4,21,and 567 are important questions"
s=re.finditer(pattern,t)
print("length 1 to 3")
for n in s:
     print(n.group(0))

```
